```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
import re
import matplotlib.pyplot as plt

# Ensure necessary NLTK corpora are downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    """
    Preprocess the input text by removing unnecessary characters,
    tokenizing, removing stop words, and lemmatizing tokens.
    """
    # Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    # Remove extra spaces and new lines
    text = re.sub(r'\s+', ' ', text)

    # Tokenize the text into words
    words = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.lower() not in stop_words]

    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    # Join the words back into a single string
    cleaned_text = ' '.join(words)
    return cleaned_text
def analyze_sentiment(text):
    """
    Analyze the sentiment of the input text using TextBlob.
    Returns the polarity and subjectivity.
    """
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Read the script file
with open('/content/drive/MyDrive/Colab Notebooks/bee movie.txt', 'r') as file:
    script = file.read()

# Preprocess the text
cleaned_script = preprocess_text(script)

# Tokenize the script into sentences
sentences = nltk.sent_tokenize(cleaned_script)

# Analyze sentiment for each sentence
results = []
for sentence in sentences:
    polarity, subjectivity = analyze_sentiment(sentence)
    results.append({
        'sentence': sentence,
        'polarity': polarity,
        'subjectivity': subjectivity
    })

# Display results
for result in results:
    print(f"Sentence: {result['sentence']}")
    print(f"Polarity: {result['polarity']}, Subjectivity: {result['subjectivity']}\n")
```

```python
import matplotlib.pyplot as plt

# Extract polarity and subjectivity for plotting
polarities = [result['polarity'] for result in results]
subjectivities = [result['subjectivity'] for result in results]

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(polarities, subjectivities, alpha=0.5)
plt.title('Sentiment Analysis of Movie Script')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.grid(True)
plt.show()
```
